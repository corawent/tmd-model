```python
tmds=['MoS_2','WS_2','MoSe_2','WSe_2']
params={}

#new parameters
#params['MoS_2']=np.array([8,np.inf,np.inf,0.008,np.inf,np.inf,3.5,0.51,0.54,0.03,1.8,-15]) #T_n, T_p 8e-13
#params['WS_2']=np.array([2,np.inf,np.inf,0.031,np.inf,np.inf,2.4,0.35,0.34,0.03,2.1,-20]) #T_n, T_p 1e-12
#params['MoSe_2']=np.array([80,26.7,np.inf,0.029,np.inf,0.3,0.03,0.50,0.58,0.03,1.5,-4]) #T_n, T_p 9e-13
#params['MoSe_2']=np.array([900,270,np.inf,0.5,np.inf,3,0.0003,0.50,0.58,0.03,1.5,-4,1e-12,1e-12]) #T_n, T_p 9e-13 TESTING
#params['WSe_2']=np.array([16.7,1.9,np.inf,0.018,np.inf,0.4,0.06,0.35,0.35,0.03,1.7,0]) #T_n, T_p 1e-12
#params['WSe_2']=np.array([25,1.9,0.33,0.2,np.inf,0.4,0.06,0.35,0.35,0.03,1.7,0]) #T_n, T_p 1e-12 - TESTING
#labels=['txr','txnr','ttr_n','ttnr_n','ttr_p','ttnr_p','Cbx','me','mh','Et','Eg','Vth']

#original parameters except wse2
params['MoS_2']=np.array([8,np.inf,0.11,0.05,np.inf,np.inf,3.5,0.51,0.54,0.03,1.8,-20,2e-11,0.0]) #their fitting, them & me (agree)
params['WS_2']=np.array([2,np.inf,0.032,0.1,np.inf,np.inf,2.4,0.35,0.34,0.03,2.1,-20,5e-12,0.0]) #their fitting, them & me (agree)
params['MoSe_2']=np.array([80,26.7,4.4,0.066,50,50,0.03,0.50,0.58,0.03,1.5,-4,3e-12,1e-10]) # their fitting, them
params['WSe_2']=np.array([16.7,1.9,0.33,0.033,1,1,0.06,0.35,0.35,0.03,1.7,0,2.5e-11,3.5e-10]) #their fitting, them
#params['WSe_2']=np.array([16.7,1.9,0.33,0.1,1,10,.06,0.35,0.35,0.03,1.7,0,2.5e-11,3.5e-10]) #their fitting, me
#labels=['txr','txnr','ttr_n','ttnr_n','ttr_p','ttnr_p','Cbx','me','mh','Et','Eg','Vth','T_n','T_p']

#BEST
#params['MoS_2']=np.array([8,np.inf,np.inf,0.008,np.inf,np.inf,3.5,0.51,0.54,0.03,1.8,-20]) #new parameters
#params['WS_2']=np.array([2,np.inf,np.inf,0.05,np.inf,np.inf,2.4,0.35,0.34,0.03,2.1,-20]) #new parameters
#params['MoSe_2']=np.array([80,26.7,4.4,0.066,50,50,0.03,0.50,0.58,0.03,1.5,-4,3e-12,1e-10]) # their fitting, them
#params['WSe_2']=np.array([25,1.9,0.33,0.2,15,150,0.03,0.35,0.35,0.03,1.7,0,2.5e-11,5e-10]) #hand-fit to their params
#labels=['txr','txnr','ttr_n','ttnr_n','ttr_p','ttnr_p','Cbx','me','mh','Et','Eg','Vth','T_n','T_p']
all_plqy={}
```

Constants

```python
k=1.381e-23 #J/K
t=300 #K
h=6.626e-34 #J/s
hbar=h/(2*np.pi)
m0=9.109e-31 #kg
q=1.602e-19 #coulombs
c=2.998e8 #m/s
eps0=8.85e-12 #F/m
```

Initialize arrays

```python
gs=np.logspace(15, 21, 100)
Vs=np.arange(-20,21,1)

plqys=np.zeros((np.size(gs),np.size(Vs)))
nxs=np.zeros((np.size(gs),np.size(Vs)))
nes=np.zeros((np.size(gs),np.size(Vs)))
nts=np.zeros((np.size(gs),np.size(Vs)))
rxrs=np.zeros((np.size(gs),np.size(Vs)))
rxnrs=np.zeros((np.size(gs),np.size(Vs)))
rtrs=np.zeros((np.size(gs),np.size(Vs)))
rtnrs=np.zeros((np.size(gs),np.size(Vs)))
rbs=np.zeros((np.size(gs),np.size(Vs)))
```

Load PLQY data

```python
plqy_data={}

for V_data in V_datas:
    key=tmds[tmd]+'_'+str(V_data)+'V'
    plqy_data=np.loadtxt(key+'.txt',skiprows=1) #change to skiprows=2 to skip artificial data point
    all_plqy[key+'_data']=plqy_data
```

TMD parameters

```python
#easier than making a dictionary - don't want to type in d every time
txr=params[tmds[tmd]][0]*1e-9 #s
txnr=params[tmds[tmd]][1]*1e-9 #s
ttr_n=params[tmds[tmd]][2]*1e-6 #s - for some reason was reported in µs
ttnr_n=params[tmds[tmd]][3]*1e-9 #s
ttr_p=params[tmds[tmd]][4]*1e-6 #s - same as above
ttnr_p=params[tmds[tmd]][5]*1e-9 #s
Cbx=params[tmds[tmd]][6]
me=params[tmds[tmd]][7]*m0
mh=params[tmds[tmd]][8]*m0
Et=params[tmds[tmd]][9]
Eg=params[tmds[tmd]][10]
Vth=params[tmds[tmd]][11]

#calculate total lifetimes for each quasiparticle
tx=1/(1/txnr+1/txr)
tt_n=1/(1/ttnr_n+1/ttr_n)
tt_p=1/(1/ttnr_p+1/ttr_p)

#calculate ni
gc2d=me/(np.pi*hbar**2)*1/(100**2) #1/cm^2*energy
gv2d=mh/(np.pi*hbar**2)*1/(100**2) #1/cm^2*energy
ni=(gc2d*gv2d)**0.5*(k*t)*np.exp(-q*Eg/(2*k*t))

#calculate T_n and T_p
if np.size(params[tmds[tmd]])==14:
    T_n=params[tmds[tmd]][12]
    T_p=params[tmds[tmd]][13]
    
elif np.size(params[tmds[tmd]])==12:
    mt_n=2*me+mh
    mt_p=me+2*mh
    mx=me+mh
    T_n=np.pi*hbar**2/(k*t)*mt_n/(me*mx)*np.exp(q*Et/(k*t))*1e4
    T_p=np.pi*hbar**2/(k*t)*mt_p/(me*mx)*np.exp(q*Et/(k*t))*1e4

#for initial N, P guesses
Cox=1/(1/(3.8*eps0*1e-2/50e-7)+1/(2*eps0*1e-2/100e-7)) #50 nm oxide=50e-7 cm; 8.85e-12F/m=8.85e-14F/cm
```

Loop over voltages and generation rates

```python
j=0
for V in Vs:
    
    #Solve for initial N, P concentrations
    N0P0=np.abs(Cox/q*(V-Vth))

    if V>Vth: #then we are electron doping, ef towards CB
        NPguess=[N0P0,ni**2/N0P0]

    elif V<Vth: #then we are hole doping, ef towards VB
        NPguess=[ni**2/N0P0,N0P0]

    elif V==Vth:
        NPguess=[ni,ni]


    def NPfunc(y):
        #x[0]=N, x[1]=P

        #net charged carriers
        out = [Cox*(V-Vth)/q-(y[0]-y[1])]

        #2D mass action law. Modified becuase of v small & v big numbers!

        if NPguess[0]>0.01*gc2d*k*t: #N is large, P is very small – Taylor expand for P portion
            out.append(gc2d*k*t*(np.exp(y[0]/(gc2d*k*t))-1)*y[1]-ni**2)

        elif NPguess[1]>0.01*gv2d*k*t: #P is large, N is very small – Taylor expand for N portion
            out.append(y[0]*gv2d*k*t*(np.exp(y[1]/(gv2d*k*t))-1)-ni**2)

        else: #not degenerately doped
            out.append(y[0]*y[1]-ni**2)      

        return out

    y0, infodict, ier, mesg = fsolve(NPfunc, NPguess, full_output=True)           

    N=y0[0]
    P=y0[1]
    
    #Solve polynomial equations to get carrier concentrations at all G values
    i=0
    for g in gs:     
        #solve for nx, rx
        #poly_nx=np.poly1d([Cbx*T_p*T_n*Cbx*T_n,Cbx+T_n/tx,1/tx+N*T_n/tt_n-g*T_n,-g])
        poly_nx=np.poly1d([Cbx*T_p*T_n,Cbx*(T_n+T_p)+T_n*T_p/tx,
                            Cbx+T_n*T_p*(-g+P/tt_p+N/tt_n)+(T_n+T_p)/tx,  
                           1/tx+N*T_n/tt_n+P*T_p/tt_p-g*(T_n+T_p),-g])
        roots=poly_nx.r
        nx=np.real(roots[np.logical_and(np.isreal(roots),roots>0)])
        rxr=nx/txr
        rxnr=nx/txnr

        #solve for ne
        ne=N/(1+T_n*nx)+P/(1+T_p*nx)

        #solve for nt
        nt_n=T_n*nx*N/(1+T_n*nx)
        nt_p=T_p*nx*P/(1+T_p*nx)
        nt=nt_n+nt_p
        
        #rates
        rtr_n=nt_n/ttr_n
        rtnr_n=nt_n/ttnr_n
        rtr_p=nt_p/ttr_p
        rtnr_p=nt_p/ttnr_p

        rtr=rtr_n+rtr_p
        rtnr=rtnr_n+rtnr_p

        #biexciton recombination rate
        rb=nx**2*Cbx

        #plqy
        plqy=1/g*(nx/txr+nt_n/ttr_n+nt_p/ttr_p)

        #put into matrices
        nxs[i,j]=nx
        nes[i,j]=ne
        nts[i,j]=nt

        rxrs[i,j]=rxr
        rxnrs[i,j]=rxnr
        rtrs[i,j]=rtr
        rtnrs[i,j]=rtnr
        rbs[i,j]=rb
        
        plqys[i,j]=plqy

        i+=1
    
    j+=1
```

### Plots

Plot PLQY contour plot

```python
mylevels=np.logspace(-4,0,51)*1e2

fig,ax=plt.subplots()
cplot=ax.contourf(Vs,gs,plqys*1e2,levels=mylevels,locator=ticker.LogLocator())
ax.set_yscale('log')
ax.set_xlabel('$V_g$ (V)')
ax.set_ylabel('Generation Rate ($cm^{-2}s^{-1}$)')
ax.set_title('$'+tmds[tmd]+'$')
cbar = fig.colorbar(cplot,ticks=[100,10,1,0.1,0.01],label='PLQY (%)')

plt.show()
fig.savefig(folder+'/'+tmds[tmd]+'_plqy_contour.png')
```

Plot slices of this plot where we have data

```python
fig,ax=plt.subplots(figsize=figuresize)

i=0
for V_data in V_datas:
    key=tmds[tmd]+'_'+str(V_data)+'V'
    plqy_plot=all_plqy[key+'_data']
    ax.loglog(plqy_plot[:,0],plqy_plot[:,1],'.',color=colors[i],label='V='+str(V_data)+'V')
    
    iplot=np.where(Vs==V_data)[0][0]
    ax.loglog(gs,plqys[:,iplot]*1e2,color=colors[i])
    i+=1

ax.set_xlabel('Generation Rate ($cm^{-2}s^{-1}$)')
ax.set_ylabel('PLQY (%)')
ax.set_ylim([1e-2,1e2])
ax.set_title('$'+tmds[tmd]+'$')
ax.legend()
fig.savefig(folder+'/'+tmds[tmd]+'_plqy_data.png')
```

Plot rates at V to plot

```python
Vtoplot=-20

iplot=np.where(Vs==Vtoplot)[0][0]
key=tmds[tmd]+'_'+str(Vtoplot)+'V'

fig,ax=plt.subplots(figsize=figuresize)
ax.axvline(x=2e17,color=[1,1,0,0.5],linewidth='3')
ax.loglog(gs,rxrs[:,iplot],color=colors[0],label='Radiative Exciton')
ax.loglog(gs,rxnrs[:,iplot],'--',color=colors[0],label='Nonradiative Exciton')
ax.loglog(gs,rtrs[:,iplot],color=colors[1],label='Radiative Trion')
ax.loglog(gs,rtnrs[:,iplot],'--',color=colors[1],label='Nonradiative Trion')
ax.loglog(gs,rbs[:,iplot],'--',color=colors[2],label='Biexciton')
ax.set_ylim([1e3,1e22])
ax.set_xlabel('Generation Rate ($cm^{-2}s^{-1}$)')
ax.set_ylabel('Recombination Rate ($cm^{-2}s^{-1}$)')
ax.set_title('$'+tmds[tmd]+'$, '+str(Vtoplot)+' V')
ax.legend(loc='best')
fig.savefig(folder+'/'+key+'_rates') #/changing-vth
```

Plot carrier concentrations at V to plot

```python
iplot=np.where(Vs==Vtoplot)[0][0]
key=tmds[tmd]+'_'+str(Vtoplot)+'V'

fig,ax=plt.subplots(figsize=figuresize)
ax.axvline(x=2e17,color=[1,1,0,0.5],linewidth='3')
ax.loglog(gs,nxs[:,iplot],label='Exciton')
ax.loglog(gs,nts[:,iplot],label='Trion')
ax.loglog(gs,nes[:,iplot],label='Electron')
ax.set_ylim([1e0,1e13])
ax.set_xlabel('Generation Rate ($cm^{-2}s^{-1}$)')
ax.set_ylabel('Concentration ($cm^{-2}$)')
ax.set_title('$'+tmds[tmd]+'$, '+str(Vtoplot)+' V')
ax.legend()
fig.savefig(folder+'/'+key+'_conc')
```
